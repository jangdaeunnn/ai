## 경사하강법으로 학습하자

```python
# 변화율은 얼마나 변해야하는지를 알려주는거고, 에러는 어디로 변해야하는지를 알려주는거야.
```

### 첫번째 샘플

```python
# 먼저, 샘플 하나에 대해서 해보자. 큰 틀의 알고리즘은 선형회귀이다.
w=1.0
b=1.0
y_hat=w*x[0]+b
print(y_hat)
```

```python
print(y[0])
```

```python
w_inc=w+0.1
y_hat_inc=w_inc*x[0]+b
w_rate=(y_hat_inc-y_hat)/(w_inc-w)
b_rate=1
print(w_rate)
```

```python
# 오차역전파를 사용하자.
err=y[0]-y_hat
w_new=w+w_rate*err
b_new=b+1*err
print(w_new,b_new)
```

### 두번째 샘플

```python
y_hat=x[1]*w_new+b_new
err=y[1]-y_hat
w_new=w_new+x[1]*err
b_new=b_new+1*err

print(w_new,b_new)
```

```python
for x_i,y_i in zip(x,y):
  y_hat=x_i*w+b
  err=y_i-y_hat
  w+=x_i*err
  b+=1*err
```

```python
print(w,b)
```

```python
plt.scatter(x,y)
pt1=(-0.1,-0.1*w+b)
pt2=(0.15,0.15*w+b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
for i in range(1,100):
  for x_i,y_i in zip(x,y):
    y_hat=x_i*w+b
    err=y_i-y_hat
    w+=x_i*err
    b+=1*err
print(w,b)
```

```python
plt.scatter(x,y)
pt1=(-0.01,-0.01*w+b)
pt2=(0.15,0.15*w+b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
