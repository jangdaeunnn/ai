## 선형회귀 뉴런으로 만들기

```python
class Neuron:
  def __init__(self):
    self.w=1.0
    self.b=1.0

  def forpass(self,x):
    y_hat=self.w*x+self.b
    return y_hat

  def backprop(self,err,x):
    w_grad=x*err
    b_grad=1.0*err
    return w_grad,b_grad

  def fit(self,x,y,epochs=100):
    for i in range(epochs):
      for x_i,y_i in zip(x,y):
        err=y_i-self.forpass(x_i)
        w_grad,b_grad=self.backprop(err,x_i)
        self.w+=w_grad
        self.b+=b_grad
    return self.w,self.b
```

## 데이터 불러오기

```python
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
```

```python
x=diabetes.data[:,2]
y=diabetes.target
```

```python
neuron=Neuron()
neuron.fit(x,y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
pt1=(-0.1,-0.1*neuron.w+neuron.b)
pt2=(0.15,0.15*neuron.w+neuron.b)
plt.plot([pt1[0],pt2[0]],[pt2[0],pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
